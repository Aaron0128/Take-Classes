```python
##整个词的代入
text_name='李白诗句.txt'
ci = '秦帝'
count = 0
with open(text_name,'r') as f:
    a = f.readline()
    while a:
        for i in range(len(a)-1):
            if a[i] == ci[0] and a[i+1] == ci[1]:
                count += 1
        #print(a)
        a = f.readline()

print(count)
```
